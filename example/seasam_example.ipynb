```python
import numpy as np
```

```python
from seasam import seasam_sample
```

## define the input parameters

note that input data should be single-precision floats (np.float32), and the output indices are long (np.int64)


```python
# requierd parameters
path_database = 'dataset.npy'
size_database = 10000
len_sequence = 128
path_train_indices = 'train_indices.npy'
size_train = 1000

# optional parameters
path_val_indices = 'validation_indices.npy'
size_val = 200
path_sorted_indices = 'sorted_indices.npy'
sax_cardinality = 8
paa_segments = 16
```

## run seasam_sample

```python
seasam_sample(path_database=path_database,
              size_database=size_database,
              len_sequence=len_sequence,
              path_train_indices=path_train_indices,
              size_train=size_train,
              path_val_indices=path_val_indices,
              size_val=size_val,
              path_sorted_indices=path_sorted_indices,
              sax_cardinality=sax_cardinality,
              paa_segments=paa_segments)
```

```python
sampled_train_indices = np.fromfile(path_train_indices, dtype=np.int64)
print(sampled_train_indices)
```

```python
# optional

sampled_val_indices = np.fromfile(path_val_indices, dtype=np.int64)
print(sampled_val_indices)
```

```python
# optional

seasam_sorted_indices = np.fromfile(path_sorted_indices, dtype=np.int64)
print(seasam_sorted_indices)
```
